```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline

#Calculate the mass of the molecules in the system
N_A = (6.022*(10**23)) #Avogadro's Number
Mr_molecule = 18.015
No_molecules = int(input("How many water molecules are in the system? "))
Mass_system = ((Mr_molecule*No_molecules)/(N_A))
print ("The overall mass of the system is ", Mass_system, "g")
```

```python
#Input the volume data of the system
volume_angstrom = pd.read_table("production_volume.dat")
for x in volume_angstrom:
    volume_cm = ((volume_angstrom)*10**-24)
    print (volume_cm[0:5])
    density = Mass_system / volume_cm
    print ("Densities calculated successfully")
    density=np.array(density)
    print (density[0:10])

print ('There are ', len(density), 'values for the Density')
```

```python
sns.set_style('darkgrid')

timestep = int(input('What is the timestep in fs? '))

xvals = []
for n in range(0,(len(density))):
    ps = n * timestep
    xvals.append(ps)

xvals=np.array(xvals)
print(xvals)
x=xvals
y=density

plt.rcParams.update({'font.size': 16})

fig= plt.figure(figsize=(10,5))
plt.xlabel('Time, ps')
plt.ylabel('Density, g/cm$^3$')
plt.plot(x,y, label='298K', linewidth=3)
plt.title('Density, Water 645')
plt.legend()
fig.savefig("density_645_water_298K.png", dpi=1000)
```
