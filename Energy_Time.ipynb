```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
timestep = int(input('What is the timestep in fs? '))
energy = pd.read_table('production_energies.dat')
y = np.array(energy)
```

```python
sns.set_style('darkgrid')

xvals = []
for n in range(0,(len(energy))):
    ps = n * timestep
    xvals.append(ps)

xvals=np.array(xvals)
print(xvals)
x=xvals

plt.rcParams.update({'font.size': 16})

fig= plt.figure(figsize=(10,5))
plt.xlabel('Time, ps')
plt.ylabel('Total Energy, kcal mol$^-1$')
plt.title('Total Energy, Water 645')
plt.plot(x,y, label='298K', linewidth=3, color='maroon')
plt.legend()
fig.savefig("energy_645_water_298K.png", dpi=1000)
```
